```python
import pandas as pd 
csv = pd.read_csv('../src/dataset/BirthRate_Parsed.csv')
exportData = {}
```

```python
keys = list(csv.keys())[3:]
keys
```

```python
for i in range(len(csv)):
    data = csv.iloc[i]
    exportData[data['Country Name'].replace(' ','_').replace(',','').replace('.','')] = [{'year' : key , 'value' : data[key]} for key in keys]
```

```python
exportData = {'BirthRate' : exportData}
exportData
```

```python
import json
```

```python
a = json.dumps(exportData)
```

```python
file = open('../src/dataset/BirthRate.json','w')
file.write(a)
```
