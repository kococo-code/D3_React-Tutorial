```python
import pandas as pd 
from glob import glob 
```

```python
csv = glob('dataset/*.csv')
```

```python
csv
```

```python
_ = pd.read_csv(csv[0])
```

```python
for i in range(len(_)):
    print(_.iloc[i])
```

```python
columns = _.keys()
```

```python
keys = list(columns[2:4])
print(keys)
```

```python
keys.extend([key[0:4] for key in list(columns[4:])])
data = []
```

```python
for i in range(len(_)):
    k = list((_.iloc[i][2:]))
    if k[0] != float('nan'):
        data.append(k)
```

```python
print(keys)
```

```python
data[-1]
```

```python
df = pd.DataFrame(columns=keys,data=data)
```
